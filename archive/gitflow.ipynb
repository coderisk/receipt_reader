---

**1. Diff View icon**

The **Diff** toggle (in the header bar) highlights messages that have changed since the last git commit. Changed messages display as syntax-highlighted unified diffs (green = added, red = removed). If you click to edit a diff-toggled message, it opens a **side-by-side Monaco diff editor** focused on the first change, where you can revert individual changes. Requires versioning to be enabled first.

---

**2. Enable Versioning**

Click the **"Enable versioning"** button in the folder view to initialize a git repo for that folder. Once enabled:
- Press **`s`** in a dialog to save + auto-commit
- Press **`Shift+C`** (Checkpoint button) to commit with a custom message
- Solveit automatically **squashes** repeated auto-save commits to keep history clean, while preserving manual checkpoint commits
