# Receipt Reader app

FastHTML + MonsterUI web app that converts PDFs/images (e.g. receipts) into editable Markdown via the [Datalab Marker API](https://www.datalab.to). Served with `JupyUvi` so it runs from inside the solveit kernel.

## Key pieces in `main.py`
- **Async API helpers**: `submit_marker` / `poll_marker` (single), `submit_markers` / `poll_markers` (batch), `convert_pdf` / `convert_pdfs` (submit + poll), `pdf2md` / `pdfs2md` (convert + save markdown & images to disk).
- **Web UI**: route `/home` — upload form on the left, preview + editable markdown panel on the right, with Copy/Reset buttons (small inline JS).
- **App objects**: `app`, `rt` (FastHTML routing).

## Conventions
- Auth via `DATALAB_KEY` env var.
- Sample receipts in folder: `sysco.jpg`, `sjd.jpg` (with matching `.md` outputs).

## Typical workflow
```
from main import *
server = JupyUvi(app)   # start
server.stop()           # stop
```
